```python
import pandas as pd
%pylab inline
import pylab as pl

pl.style.use('bmh')
pl.rcParams["figure.figsize"] = (10,10)
pl.rcParams["font.size"] = 25
pl.rcParams["xtick.labelsize"] = 16
pl.rcParams["ytick.labelsize"] = 16
```

```python
imstats = pd.read_csv("plumes_finder_save.csv")
PLUM0 = 143
PLUM1 = 149

def plotPlumStats(df, x, y, cut=False):
    #print (df[x])
    if cut: 
        cut0 = PLUM0 - 5
        cut1 = PLUM1 + 5 
    else:
        cut0 = 0
        cut1 = -1
    pl.plot(df[x][cut0:cut1], df[y][cut0:cut1], '.', ms = 10, alpha=0.5, label="plume")
    pl.plot(df[x][PLUM0:PLUM1], df[y][PLUM0:PLUM1], '.', ms = 10, alpha=0.5, label="no plume")
    pl.legend()
```

```python
imstats.head()
```

```python
pl.figure()
pl.xlabel("chi2 mean subtracted")
pl.ylabel("chi2 median subtracted")
plotPlumStats(imstats, 'chisq_mn', 'chisq_md', cut=True)
pl.figure()
pl.xlabel("standard deviation of mean subtracted")
pl.ylabel("standard deviation of median subtracted")
plotPlumStats(imstats, 'std_mn', 'std_md', cut=True)
pl.figure()
pl.xlabel("mean of mean subtracted")
pl.ylabel("mean of median subtracted")
plotPlumStats(imstats, 'mean_mn', 'mean_md', cut=True)
```

```python
pl.figure()
pl.xlabel("chi2 mean subtracted")
pl.ylabel("chi2 median subtracted")
plotPlumStats(imstats, 'chisq_mn', 'chisq_md')
pl.figure()
pl.xlabel("standard deviation of mean subtracted")
pl.ylabel("standard deviation of median subtracted")
plotPlumStats(imstats, 'std_mn', 'std_md')
pl.figure()
pl.xlabel("mean of mean subtracted")
pl.ylabel("mean of median subtracted")
plotPlumStats(imstats, 'mean_mn', 'mean_md')
```
